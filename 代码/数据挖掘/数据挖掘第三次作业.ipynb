```python
# coding=utf-8
import pandas as pd
```

```python
# 读取Excel文件
df = pd.read_excel('D:\\大三上\\数据挖掘与商务智能\\第三次实验课20221016\\实验数据(xlsx).xlsx')
#创建标题行
df.columns = ['A', 'B', 'C', 'D','E','F','G','H','I','J']
```

1

```python
# 处理表格属性,删除无意义的列
df1 = df.drop(labels='A', axis=1)
df2 = df1.drop(labels='H', axis=1)
df3 = df2.drop(labels='I', axis=1)
print (df3)
```

2

```python
df[['B','E','F','D']] = df[['B','E','F','D']].astype(str)
# 查找特定的数据
result = df[(df['B'] == '东城区') & (df['E'] == '2016-01-06') & (df['F'] == '09:00:00')]['D']
print(result)
```

3

```python
# 将拥堵状态替换为数值
# 用于将拥堵状态转换为数值的映射字典
mapping_dict = {"畅通": 0, "基本畅通": 1, "轻度拥堵": 2, "中度拥堵": 3, "严重拥堵": 4}
```

```python
# 将交通拥堵状态替换为对应的数值型
df['D'] = df['D'].map(mapping_dict).fillna(0)
print (df)
```
